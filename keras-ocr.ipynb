```python
import matplotlib.pyplot as plt

import keras_ocr
```

```python
pipeline = keras_ocr.pipeline.Pipeline()
```

```python
images = [keras_ocr.tools.read(url) for url in [
        'https://dynaimage.cdn.cnn.com/cnn/q_auto,w_1903,c_fill,g_auto,h_1070,ar_16:9/http%3A%2F%2Fcdn.cnn.com%2Fcnnnext%2Fdam%2Fassets%2F180927155906-for-freedoms-billboard.jpg',
        'https://dynaimage.cdn.cnn.com/cnn/q_auto,w_727,c_fit/http%3A%2F%2Fcdn.cnn.com%2Fcnnnext%2Fdam%2Fassets%2F180906114659-for-freedoms-9.jpg',
        'https://dynaimage.cdn.cnn.com/cnn/q_auto,w_727,c_fit/http%3A%2F%2Fcdn.cnn.com%2Fcnnnext%2Fdam%2Fassets%2F180924111633-for-freedoms-12.jpg'
    ]
]
```

```python
import numpy as np
print(np.shape(images)) 
```

```python
prediction_groups = pipeline.recognize(images)

# Plot the predictions
fig, axs = plt.subplots(nrows=len(images), figsize=(20, 20))
for ax, image, predictions in zip(axs, images, prediction_groups):
    keras_ocr.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)
```
